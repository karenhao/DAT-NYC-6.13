### Problem

1. Draw a scatter plot of the houses location, coloring them based on the median house value.
1. Transform the tail heavy distributions using mathematical operations.
2. Create three new columns:
   1. `rooms_per_household`
   2. `bedrooms_per_room`
   3. `population_per_household`
   
3. Drop values where na's are missing, assign this DataFrame to `dropped_houses`.
4. Replace missing values with mean or median values, assign this to `replaced_houses`.

#### draw scatter of houses' locations

```python
plt.figure(figsize=(9,10))
plt.scatter(df.latitude,df.longitude,alpha=0.3,c=df.median_house_value)
plt.colorbar()
sns.despine(left=True,bottom=True)
```

#### transform the tail heavy distributions using mathematical operations

```python
plt.hist(np.log( df['total_rooms']), bins = 30);
print('The skew before was', df['total_rooms'].skew(), '\nAfter transforming is', np.log(df['total_rooms'].skew()))
sns.despine(left=True,bottom=True)
```

```python
plt.hist(np.log( df['households']), bins = 30);
print('The skew before was', df['households'].skew(), '\nAfter transforming is', np.log(df['households'].skew()))
sns.despine(left=True,bottom=True)
```

```python
plt.hist(np.log( df['population']), bins = 30);
print('The skew before was', df['population'].skew(), '\nAfter transforming is', np.log(df['population'].skew()))
sns.despine(left=True,bottom=True)
```

```python
plt.hist(np.log( df['median_income']), bins = 30);
print('The skew before was', df['median_income'].skew(), '\nAfter transforming is', np.log(df['median_income'].skew()))
sns.despine(left=True,bottom=True)
```

#### create new columns

```python
df['rooms_per_household'] = df['total_rooms']/df['households']
df['bedrooms_per_room'] = df['total_bedrooms']/df['total_rooms']
df['population_per_household'] = df['population']/df['households']
```

```python
# correlation matrix
## note: diagonal correlation = 1
corr = df.corr()

corr
```

```python
corr['median_house_value']
```

#### drop values where na's are missing

```python
dropped_houses = df.dropna()
```

```python
dropped_houses.head()
```

#### replace missing values with mean or median values

```python
replaced_houses = df.fillna(df.mean())
```
